Chosen paper: [Luo, J., Jia, Y., & Liu, X. (2022). Understanding NFT price moves through Social media keywords analysis. *arXiv preprint arXiv:2209.07706*](https://arxiv.org/abs/2209.07706)

Relation to Moonbirds: the Moonbirds was involved as one kind of NFT in this paper.

Background/Motivation: Non-Fungible Token (NFT) is evolving with the rise of the cryptocurrency market and the development of blockchain techniques,
which leads to an emerging NFT market that has become prosperous rapidly. However, the overall rise procedure of the NFT market has not
been well understood.This paper aims to fill this gap.

Research Questions: The first research question is does the activeness of
a social media community help in forecasting the prices and or
reversely. The second research question is will the causal relationship be inspected stronger in authentic NFT projects than the copycat projects since copycat project may not have sustained social media community building

Application Scenario: It is applied to the NFT market prices scenarios (including Moonbirds). Results: the result shows that social media words as the predictors result in all 19 top projects having a testing accuracy above the random baseline.
